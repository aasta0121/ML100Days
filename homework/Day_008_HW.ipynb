```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd

# 設定 data_path
dir_data = './data/'
```

```python
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
print(app_train['AMT_INCOME_TOTAL'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
print(f"總收入平均數:{app_train['AMT_INCOME_TOTAL'].mean()}")
print(f"總收入標準差:{app_train['AMT_INCOME_TOTAL'].std()}")

income_type_df = app_train[['AMT_INCOME_TOTAL','NAME_INCOME_TYPE']]
income_type_df.rename(columns={'AMT_INCOME_TOTAL':'Total Income','NAME_INCOME_TYPE':'Income type'}, inplace=True)
mean_type_df = income_type_df.groupby('Income type').mean()
mean_type_df.plot.bar(rot=45)

income_edu_df = app_train[['AMT_INCOME_TOTAL','NAME_EDUCATION_TYPE']]
income_edu_df.rename(columns={'AMT_INCOME_TOTAL':'Total Income','NAME_EDUCATION_TYPE':'Education'}, inplace=True)
mean_edu_df = income_edu_df.groupby('Education').mean()
mean_edu_df.plot.bar(rot=45)
```
